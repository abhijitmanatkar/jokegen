```python
import matplotlib.pyplot as plt
```

```python
with open('./loss.txt', 'r') as f:
    lines = f.readlines()
```

```python
itr = []
train_loss = []
val_loss = []

for line in lines:
    i, t, v = line.split()
    itr.append(int(i))
    train_loss.append(float(t))
    val_loss.append(float(v))
```

```python
plt.plot(itr, train_loss)
plt.plot(itr, val_loss)
plt.xlabel("Iterations")
plt.ylabel("Losses")
plt.legend(["Train loss", "Validation loss"])
```
